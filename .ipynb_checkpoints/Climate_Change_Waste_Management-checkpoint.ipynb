```python
wdi = pd.read_csv('Data.csv')
```

```python
wdi = wdi.drop(['Series Code'],axis=1)
```

```python
wdi = wdi[wdi['Country Name'].notna()] #drop missing country rows
```

```python
wdi = wdi[wdi['Series Name'].notna()] #drop missing series/variable names
```

```python
cc_nan = wdi[wdi['Country Code'].isna()]
cc_nan
```

```python
#Create function to clean wdi- by reshaping into 1 time variable. 
#Merge oecd and wdi and drop whatever is not common. See alternatively if you can merge all keep uncommon vars too. 
#will have to rename oecd's cou as Country Code
```

```python
wdi_clean = wdi.rename(columns=lambda x: re.sub(r'^\d+ \[(YR\d+)\]$',r'\1',x))
```

```python
wdi_clean.head()
```

```python
wdi_reshaped = pd.wide_to_long(wdi_clean.reset_index(), stubnames=["YR"], i=["index","Country Name","Country Code"], j="Year",\
                               suffix='\d+')
```

```python
wdi_reshaped = wdi_reshaped.reset_index().reindex(["Country Code", "Country Name", "Year", "Series Name", "YR"], axis=1)
```

```python
wdi_reshaped.head()
```

```python
wdi_reshaped['Series Name'].unique()
```

```python
wdi_final = pd.pivot(wdi_reshaped, index=['Year','Country Code','Country Name'],columns='Series Name',values='YR')
```

```python
wdi_final = wdi_final.replace("..", np.nan)
wdi_final.head()
```

```python
wdi_final.sort_values(by=['Country Name', 'Year'],ascending=[True, True])
```
