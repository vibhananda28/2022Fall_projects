### Ques.11& 12: Influenza Transmission Zones Tables and Plots

```python
df3 = df[['Flu Zone','Year','Week','Positive Flu']] #Subsetting to have only relevant columns. 
```

```python
df3.head()
```

```python
def zone_totals_per_week(table, zone):
    #Extracts rows belonging to specified Zone, then group by year+week and calculates the sum of 'Total Positive'. Alters the
    #derivative DataFrame's index and organization to allow plotting.
    # :param table: dataframe name.
    # :param zone: the name of an "Influenza Transmission Zone.
    
    df3 = table[table['Flu Zone'] == zone] 
    df3 = df3.groupby(['Year','Week']).sum('Positive Flu').reset_index()
    fig, ax = plt.subplots(figsize=(8,5))
    for i in np.unique(df3['Year']):
        df4 = df3.loc[df3['Year']==i]
        x= df4['Week']
        y= df4['Positive Flu']
        plt.plot(x, y, label = i)
    plt.legend()
    plt.xlabel("Weeks")
    plt.ylabel("NO. OF CASES")
    plt.title(zone.upper()+' FLU CASES BY WEEK')
    plt.show()
```

```python
# Iteration to create flu cases plots for all the zones in the dataframe. 
for x in np.unique(df3['Flu Zone']):
    zone_totals_per_week(df3, x)
```

#### Ques.3: Data Analysis: Positive Type A and Positive Type B plots

```python
df_new = df.set_index(['Year', 'Week']) #Setting Year and Week as an index
```

```python
def plot_a_flu_rate(country):
    #The function creates flu positivity rate plots for the respective country.
    #:param country: country to be plotted for. 
    
    df_new_a[df_new_a['Country'] == country].plot(title=country.upper()+' POSITIVE TYPE-A FLU CASES BY WEEK', color='red',rot=40, figsize=(8,5))
    plt.xlabel("(YEAR, WEEK)")
    plt.ylabel("PERCENT")
    plt.show()
```

```python
def plot_b_flu_rate(country):
    #The function creates flu positivity rate plots for the respective country.
    #:param country: country to be plotted for. 
    
    df_new_b[df_new_a['Country'] == country].plot(title=country.upper()+' POSITIVE TYPE-A FLU CASES BY WEEK', color='red',rot=40, figsize=(8,5))
    plt.xlabel("(YEAR, WEEK)")
    plt.ylabel("PERCENT")
    plt.show()
```
